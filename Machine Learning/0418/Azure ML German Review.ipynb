```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn import neighbors
from sklearn import datasets
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.ensemble import VotingClassifier
```

```python
data=pd.read_csv("german.csv", names=['Status of checking account','Duration in months','Credit history','Purpose', 'Credit amount', 'Savings account/bond', 'Present employment since', 'Installment rate in percentage of disposable income', 'Personal status and sex', 'Other debtors', 'Present residence since', 'Property', 'Age in years', 'Other installment plans', 'Housing', 'Number of existing credits', 'Job', 'Number of people providing maintenance for', 'Telephone', 'Foreign worker', 'Credit risk'])
#data.columns = ['Status of checking account','Duration in months','Credit history','Purpose', 'Credit amount', 'Savings account/bond', 'Present employment since', 'Installment rate in percentage of disposable income', 'Personal status and sex', 'Other debtors', 'Present residence since', 'Property', 'Age in years', 'Other installment plans', 'Housing', 'Number of existing credits', 'Job', 'Number of people providing maintenance for', 'Telephone', 'Foreign worker', 'Credit risk']
data.head()
```

```python
data['Status of checking account'],class_names=pd.factorize(data['Status of checking account'])
data['Credit history'],_=pd.factorize(data['Credit history'])
data['Purpose'],_=pd.factorize(data['Purpose'])
data['Savings account/bond'],_=pd.factorize(data['Savings account/bond'])
data['Present employment since'],_=pd.factorize(data['Present employment since'])
data['Personal status and sex'],_=pd.factorize(data['Personal status and sex'])
data['Other debtors'],_=pd.factorize(data['Other debtors'])
data['Property'],_=pd.factorize(data['Property'])
data['Other installment plans'],_=pd.factorize(data['Other installment plans'])
data['Housing'],_=pd.factorize(data['Housing'])
data['Job'],_=pd.factorize(data['Job'])
data['Telephone'],_=pd.factorize(data['Telephone'])
data['Foreign worker'],_=pd.factorize(data['Foreign worker'])
data2=data.fillna(value=0)
```

```python
x=data2.iloc[:,:-1].values
y=data2.iloc[:,-1:].values.reshape(-1,1)
x,y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
LR=LogisticRegression()
LR.fit(xtrain,ytrain)
LResult=LR.predict(xtest)
metrics.accuracy_score(ytest,LResult)
```

```python
SVM=svm.SVC(kernel='linear')
SVM.fit(xtrain,ytrain)
SResult=SVM.predict(xtest)
metrics.accuracy_score(ytest,SResult)
```

```python
DT=DecisionTreeClassifier()
DT.fit(xtrain,ytrain)
DResult=DT.predict(xtest)
metrics.accuracy_score(ytest,DResult)
```

```python
RF=RandomForestClassifier()
RF.fit(xtrain,ytrain)
RResult=RF.predict(xtest)
metrics.accuracy_score(ytest,RResult)
```

```python
KNN=KNeighborsClassifier()
KNN.fit(xtrain,ytrain)
KResult=KNN.predict(xtest)
metrics.accuracy_score(ytest,KResult)
```

```python
classifiers=([['A',LR],['B',SVM],['C',KNN],['D',DT],['E',RF]])
```

```python
for cf_name, cf in classifiers:
    cf.fit(xtrain,ytrain)
    TResult=cf.predict(xtest)
    TAccuracy=metrics.accuracy_score(ytest,TResult)
    print(TAccuracy)
```

```python
VC=VotingClassifier(estimators=classifiers)
VC.fit(xtrain,ytrain)
VResult=VC.predict(xtest)
metrics.accuracy_score(ytest,VResult)
```
