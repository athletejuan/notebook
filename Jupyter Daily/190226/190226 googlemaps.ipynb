```python
crime_seoul=pd.read_csv("crime_Seoul2.csv", encoding='utf-8', index_col=0)
station_addresses=[]
station_name=[]
station_lat=[]
station_lng=[]

for name in crime_seoul['관서명']:
    nm='서울'+name[:-1]+'경찰서'
    #print(nm)
    tmp=gmaps.geocode(nm,language='ko')
    address=tmp[0].get("formatted_address")
    tmp_loc=tmp[0].get('geometry')
    lat=tmp_loc['location']['lat']
    lng=tmp_loc['location']['lng']
    station_addresses.append(address)
    station_name.append(name)
    station_lat.append(lat)
    station_lng.append(lng)
    #print(nm,',',address,',',lat,',',lng,',',name)
print(len(station_addresses), len(station_name), len(station_lat), len(station_lng))

df=pd.DataFrame()
df['address']=station_addresses
df['name']=station_name
df['lat']=station_lat
df['lng']=station_lng

import folium
fm=folium.Map(location=[df['lat'].mean(), df['lng'].mean()],zoom_start=11)
for i in tqdm_notebook(df.index):
    folium.Marker(location=[df['lat'][i],df['lng'][i]]).add_to(fm)
'''
#print(stations)    
#for station in stations:
#    tmp=gmaps.geocode(station,language='ko')
#    print(tmp)

for station in stations:
    tmp=gmaps.geocode(station,language='ko')
#    print(tmp)
    
for i in range(len(stations)):    
    tmp2=gmaps.geocode([i], language='ko')
    #tmp2.append(tmp[i].get('geometry'))
    #lat=tmp2['location']['lat']
    #lng=tmp2['location']['lng']
    #print(tmp2['location']['lat'],tmp2['location']['lng'])
    #print(tmp2)
tmp3=tmp2[i].get('geometry')
lat=tmp3['location']['lat']
lng=tmp3['location']['lng']
print(address, tmp3['location']['lat'],tmp3['location']['lng'])
'''
'''    
for t in tmp:
    tmp2.append(t[:])
    print(tmp2)
    
    #address=tmp[:].get('formatted_address')
'''
'''
    tmp2=tmp.geometry
    lat=tmp2['location']['lat']
    lng=tmp2['location']['lng']
    print(address, tmp2['location']['lat'],tmp2['location']['lng'])
'''
'''
tmp2=gmaps.geocode('서울중부경찰서',language='ko')
address=tmp2[0].get('formatted_address')
tmp3=tmp2[0].get('geometry')
lat=tmp3['location']['lat']
lng=tmp3['location']['lng']
print(address, tmp3['location']['lat'],tmp3['location']['lng'])
'''
```
