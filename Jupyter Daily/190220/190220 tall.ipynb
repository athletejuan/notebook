```python
# data.go.kr 건강검진결과중에서 키를 구간별 나타내는 로직을 구현하고 
# matplotlib로 plot을 구현하시오.
import numpy as np
data = np.loadtxt(fname="NHIS_OPEN_GJ_2016.csv", dtype=np.str,
                 delimiter=",", usecols=range(0,7))
count1 = 0 
count2 = 0
count3 = 0 
count4 = 0 
count5 = 0
count6 = 0
for row in data[1:,5]:
    if row != '':
        if int(row) = 150:
            count1 += 1
        if int(row) = 155:
            count2 += 1
        if int(row) = 160:
            count3 += 1
        if int(row) = 165:
            count4 += 1
        if int(row) = 170:
            count5 += 1
        if int(row) = 175:
            count6 += 1
            
result = [count1, count2, count3, count4, count5, count6]
title = ['150~155', '155~ 160', '160~165', '165~170', 
         '170~175', '175~180']
import matplotlib.pyplot as plt
plt.plot(result)
plt.show()

```
