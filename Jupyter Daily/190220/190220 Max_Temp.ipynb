```python
import csv
f=open('seoul.csv',encoding='utf-8')
data=csv.reader(f)
next(data)
max_temp=0
max_temp_data=''
temp=0
temp_list=[]
for row in data:
    if row[-1]!='':
        temp+=float(row[-1])
        temp_list.append(row[-1])
        
print(temp) #최고온도의 총합
print(len(temp_list)) #자료 갯수
print(temp/len(temp_list)) #평균 최고온도
    
```

```python
import csv
f=open('seoul.csv',encoding='utf-8')
data=csv.reader(f)
next(data)
max_temp=0
max_temp_data=''
for row in data:
    if row[-1]!='':
        if(int(row[0].split("-")[0]) < 1995 and float(row[-1]) > max_temp):            
            max_temp = float(row[-1])
            max_temp_date = row[0]
            
print(max_temp, max_temp_date) #1995년 이전 최고 온도, 최고 온도인 날    
```

```python
import csv
f=open('seoul.csv',encoding='utf-8')
data=csv.reader(f)
next(data)
temperature=[]
for row in data:
    if(row[-1]) != ''and (row[0].split("-")[1] == "11" and row[0].split("-")[2] == "24"):
        temperature.append(float(row[-1]))
    
import matplotlib.pyplot as plt
plt.plot(temperature)
plt.show()
```

```python
import csv
f=open('seoul.csv',encoding='utf-8')
data=csv.reader(f)
next(data)
#역대 서울 최고 기온의 히스토그램 출력
max=[]
for row in data:
    if(row[-1]!=''):
        max.append(float(row[-1]))
        
import matplotlib.pyplot as plt
plt.hist(max, bins=100)
plt.show()
```

```python
import csv
f=open('village_population.csv',encoding='utf-8')
data=csv.reader(f)
next(data)
data2=list(data)
result=[]
for row in data2:
    if "송도1동" in row[0]:
        for i in range(3,len(row)):
            result.append(int(row[i]))
result2=[]
for row in data2:
    if "대학동" in row[0]:
        for i in range(3,len(row)):
            result2.append(int(row[i]))

import matplotlib.pyplot as plt
plt.plot(result)
plt.plot(result2)
plt.show()
```
