```pyspark3
Test_Data = spark.sparkContext.textFile('wasb:///Test_Data')
```

```pyspark3
Temp_Data = Test_Data.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda a,b:a+b)
```

```pyspark3
Temp_Data.sortBy(lambda k: k[1], False).take(5)
```
